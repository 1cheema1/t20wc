```python
import pandas as pd
import csv
df = pd.read_csv("data/t20i_matches_clean.csv")

pakWinsInOvalAndToss = df[
    (df["toss_winner"] == "Pakistan") &
    (df["winner"] == "Pakistan")&
    (df['venue']== 'Kennington Oval')]

pakplaysInOval=df[
    ((df["team1"] == "Pakistan") |
     (df["team2"] == "Pakistan")) &
    (df['venue']== 'Kennington Oval')]

pakWinsInOvalnotToss = df[
    (df["toss_winner"]  != "Pakistan") &
    (df["winner"] == "Pakistan")&
    (df['venue']== 'Kennington Oval')]

pak_plays_in_oval=len(pakplaysInOval)

pwandt=len(pakWinsInOvalAndToss)
pwandt2=len(pakWinsInOvalnotToss)


final=round(pwandt/pak_plays_in_oval * 100, 1)
final2=round(pwandt2/pak_plays_in_oval * 100, 1)
print('Pakistan has won '+ str(final) +'% matches in which it won the toss, in the Oval')
print('Pakistan has won '+ str(final2) +'% matches in which it loss the toss, in the Oval')
print(pak_plays_in_oval)
```
