```python
# We will define a toss win as a team winning both the toss AND the match.
## This cell will compute tosses.

import pandas as pd

df = pd.read_csv("data/t20i_matches_clean.csv")

total_matches=len(df)

match_and_toss=len(df[(df['toss_winner'] == df['winner'])])

toss_pct=round((match_and_toss / total_matches)*100 , 3)

print('The percent of matches won by the team who won the toss is '+str(toss_pct) +'% overall')
```

no significant effect

```python
##This cell will compute whether toss decision has an effect

bat=len(df[(df['toss_decision'] == 'bat')])
ball=len(df[(df['toss_decision'] == 'field')])

bat_win=len(df[
    (df['toss_decision'] == 'bat') &
    (df['toss_winner'] == df['winner'])  
               ])
ball_win=len(df[
    (df['toss_decision'] == 'field') &
    (df['toss_winner'] == df['winner'])  
               ])

bat_win_pct=round((bat_win/ bat )*100 , 3)
ball_win_pct=round((ball_win/ ball )*100 , 3)

print('The percentage of matches won by the team who won the toss and elected to bat is '+str(bat_win_pct)+'%')
print('The percentage of matches won by the team who won the toss and elected to field is '+str(ball_win_pct)+'%')
```

no significant effect 

```python
#This one will look at venues
# add helper column: did toss winner also win match?
df["toss_win"] = (df["toss_winner"] == df["winner"])

# group by venue
venue_stats = (
    df.groupby("venue")
      .agg(
          matches=("venue", "count"),
          toss_win_rate=("toss_win", "mean")
      )
      .reset_index()
)

# convert rate to percentage
venue_stats["toss_win_rate"] = venue_stats["toss_win_rate"] * 100

# drop small sample venues
venue_stats = venue_stats[venue_stats["matches"] >= 10]

# sort by toss impact
venue_stats = venue_stats.sort_values(by="toss_win_rate", ascending=False)

venue_stats.head(100)
```

```python
#we will count matches played for each team

played = {}

for col, row in df.iterrows():
    team1 = row["team1"]
    team2 = row["team2"]

    if team1 not in played:
        played[team1] = 0
    played[team1] += 1

    if team2 not in played:
        played[team2] = 0
    played[team2] += 1

print(played)

#count wins

won={}
for col, row in df.iterrows():
    team = row["winner"]

    if team not in won:
        won[team] = 0
    won[team] += 1

for team in played:
    if team not in won:
        won[team]=0
    
#compute win percent
wpct={}
for team in played:
    if team not in won:
        wpct[team]=0
    else:
        wpct[team]=(won[team] / played[team]) 

wpct
```

```python
##compile the 3 dicts into one dataframe
rows=[]
for team in played:
    r = {
    "Team": team,
    "Appearances": played[team],
    "Wins": won[team],
    "Win Rate": wpct[team]
}
    rows.append(r)

df1=pd.DataFrame(rows)
df1
```

```python
#sort dataframe for readibility

df1=df1.sort_values(by='Win Rate', ascending=False)

df1.head(40)
```

```python
matchupnum = {}
h2h_wins = {}

for i, row in df.iterrows():
    team1 = row["team1"]
    team2 = row["team2"]
    winner = row["winner"]

    matchup = tuple(sorted([team1, team2]))

    if matchup not in matchupnum:
        matchupnum[matchup] = 0
    matchupnum[matchup] += 1

    if matchup not in h2h_wins:
        h2h_wins[matchup] = {
            matchup[0]: 0,
            matchup[1]: 0
        }

    if winner in h2h_wins[matchup]:
        h2h_wins[matchup][winner] += 1


```

```python
#now we want to see the win rate for each h2h
h2h_rates = {}

for matchup in matchupnum:
    total_matches = matchupnum[matchup]
    team_a, team_b = matchup

    wins_a = h2h_wins[matchup][team_a]
    wins_b = h2h_wins[matchup][team_b]

    h2h_rates[matchup] = {
        team_a: wins_a / total_matches,
        team_b: wins_b / total_matches
    }
h2h_rates
```
