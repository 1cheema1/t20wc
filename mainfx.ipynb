```python
# We will define a toss win as a team winning both the toss AND the match.
## This cell will compute tosses.

import pandas as pd

df = pd.read_csv("data/t20i_matches_clean.csv")

total_matches=len(df)

match_and_toss=len(df[(df['toss_winner'] == df['winner'])])

toss_pct=round((match_and_toss / total_matches)*100 , 3)

print('The percent of matches won by the team who won the toss is '+str(toss_pct) +'% overall')
```

no significant effect

```python
##This cell will compute whether toss decision has an effect

bat=len(df[(df['toss_decision'] == 'bat')])
ball=len(df[(df['toss_decision'] == 'field')])

bat_win=len(df[
    (df['toss_decision'] == 'bat') &
    (df['toss_winner'] == df['winner'])  
               ])
ball_win=len(df[
    (df['toss_decision'] == 'field') &
    (df['toss_winner'] == df['winner'])  
               ])

bat_win_pct=round((bat_win/ bat )*100 , 3)
ball_win_pct=round((ball_win/ ball )*100 , 3)

print('The percentage of matches won by the team who won the toss and elected to bat is '+str(bat_win_pct)+'%')
print('The percentage of matches won by the team who won the toss and elected to field is '+str(ball_win_pct)+'%')
```

no significant effect 

```python
#This one will look at venues
# add helper column: did toss winner also win match?
df["toss_win"] = (df["toss_winner"] == df["winner"])

# group by venue
venue_stats = (
    df.groupby("venue")
      .agg(
          matches=("venue", "count"),
          toss_win_rate=("toss_win", "mean")
      )
      .reset_index()
)

# convert rate to percentage
venue_stats["toss_win_rate"] = venue_stats["toss_win_rate"] * 100

# drop small sample venues
venue_stats = venue_stats[venue_stats["matches"] >= 10]

# sort by toss impact
venue_stats = venue_stats.sort_values(by="toss_win_rate", ascending=False)

venue_stats.head(100)
```
