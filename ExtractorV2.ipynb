```python
import csv
import pandas as pd

def parse_info_file(filepath):
    teams = []
    match_data = {
        "winner": None,
        "outcome": None
    }

    with open(filepath, "r", encoding="utf-8") as f:
        reader = csv.reader(f)
        for row in reader:
            if len(row) < 3:
                continue

            key = row[1]
            value = row[2]

            if key == "team":
                teams.append(value)

            elif key == "date":
                match_data["date"] = value

            elif key == "venue":
                match_data["venue"] = value

            elif key == "toss_winner":
                match_data["toss_winner"] = value

            elif key == "toss_decision":
                match_data["toss_decision"] = value

            elif key == "winner":
                match_data["winner"] = value

            elif key == "outcome":
                match_data["outcome"] = value

    # assign teams
    if len(teams) == 2:
        match_data["team1"] = teams[0]
        match_data["team2"] = teams[1]
    else:
        match_data["team1"] = None
        match_data["team2"] = None

    return match_data

import os

DATA_DIR = "data/t20s_male_csv2"
files = os.listdir(DATA_DIR)
len(files)
info_files = []

for f in files:
    if f.endswith("_info.csv"):
        info_files.append(f)

len(info_files)
sample_path = os.path.join(DATA_DIR, info_files[0])

data = parse_info_file(sample_path)
print(data)
all_matches = []

for filename in info_files:
    path = os.path.join(DATA_DIR, filename)
    match = parse_info_file(path)

    # skip if no winner
    if match["winner"] is None:
        continue

    # skip malformed files
    if match["team1"] is None or match["team2"] is None:
        continue

    all_matches.append(match)

len(all_matches)
df = pd.DataFrame(all_matches)

df["winner"].isna().sum()        # should be 0
df["outcome"].value_counts()     # optional, informational
len(df)

df.to_csv("data/t20i_matches_clean.csv", index=False)

# 1. convert date column to actual datetime
df["date"] = pd.to_datetime(df["date"])

# 2. sort by date (oldest → newest)
df = df.sort_values(by="date")

# OPTIONAL: if you want newest → oldest instead, use:
# df = df.sort_values(by="date", ascending=False)

# 3. quick sanity check
print(df[["date", "team1", "team2", "winner"]].head())
print(df[["date", "team1", "team2", "winner"]].tail())

# 4. save sorted dataframe to csv
df.to_csv("data/t20i_matches_clean.csv", index=False)

print("sorted csv saved to data/t20i_matches_clean.csv")
```
