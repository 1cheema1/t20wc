```python
import pandas as pd
import csv
df = pd.read_csv("data/t20i_matches_clean.csv")

pak_toss_and_match_wins = df[
    (df["toss_winner"] == "Pakistan") &
    (df["winner"] == "Pakistan")]

pak_wins=len(df[df['winner']=='Pakistan'])

times=len(pak_toss_and_match_wins)

final=round(times/pak_wins * 100, 1)

print('Pakistan has won '+ str(final) +'% matches in which it won the toss')
```
